```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Coordinates and Tax/Census info

```python
df = pd.read_csv('Geo-Coordinates from df_final(1).csv')
df2 = pd.read_csv('df_final.csv')
```

# Including CityTax is df

```python
df['CityTax'] = df2.CityTax
```

```python
df.head()
```

# Cast as floats to visualize

```python
df.Long = df.Long.astype(float)
df.Lat = df.Lat.astype(float)
df.CityTax = df.CityTax.astype(float)
```

# Plotting the Lat/Long to see what it looks like

```python
df.plot(kind="scatter", x="Lat", y="Long", alpha=0.4)
plt.show()
```

# Now let's add some color

```python
df.plot(kind="scatter", x="Lat", y="Long",
    c="CityTax", cmap=plt.get_cmap("jet"),
    colorbar=True, alpha=0.4, figsize=(10,7),
)
plt.show()
```

# Forgot to filter out the big money properties

```python
df = df[df.CityTax <= 6514]
```

# That's better. Look at that color

# We can see areas where the property tax is high and low

```python
df.plot(kind="scatter", x="Lat", y="Long",
    c="CityTax", cmap=plt.get_cmap("jet"),
    colorbar=True, alpha=0.4, figsize=(10,7),
)
plt.show()
```

```python
df['paa10'] = df2.paa10
df['pwhite10'] = df2.pwhite10
```

# The "heat" on this map is population of African American

```python
df.plot(kind="scatter", x="Lat", y="Long",
    c="paa10", cmap=plt.get_cmap("jet"),
    colorbar=True, alpha=0.4, figsize=(10,7),
)
plt.show()
```

# The "heat" on this map is population of white

```python
df.plot(kind="scatter", x="Lat", y="Long",
    c="pwhite10", cmap=plt.get_cmap("jet"),
    colorbar=True, alpha=0.4, figsize=(10,7),
)
plt.show()
```
