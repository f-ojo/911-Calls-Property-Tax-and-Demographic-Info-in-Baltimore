```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df_final.csv')
```

```python
df.head(10)
```

```python
df2 = pd.DataFrame(columns=['Neighborhood', 'CityTaxAverage', 'paa10', 'pwhite10', 'pasi10',
                            'p2more10','ppac10', 'phisp10'])
```

```python
for i in df.Neighborhood.unique():
    tempdf = df[df.Neighborhood == i]
    average = tempdf.CityTax.mean(axis = 0)
    aa = tempdf.paa10.unique()[0]
    white = tempdf.pwhite10.unique()[0]
    asi = tempdf.pasi10.unique()[0]
    more = tempdf.p2more10.unique()[0]
    pac = tempdf.ppac10.unique()[0]
    hisp = tempdf.phisp10.unique()[0]
    
    tempdf = pd.DataFrame({"Neighborhood":[i],
                           "CityTaxAverage":[average],
                           "paa10":[aa],
                           "pwhite10":[white],
                           "pasi10":[asi],
                           "p2more10":[more],
                           "ppac10":[pac],
                           "phisp10":[hisp]})
    df2 = df2.append(tempdf)
```

```python
df2.to_csv("Neighborhood_TaxAVG_Census.csv", index = False)
```
