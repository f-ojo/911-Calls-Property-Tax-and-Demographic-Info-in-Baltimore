```python
import folium
from folium import plugins
from folium.plugins import HeatMap
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('df_final.csv')
```

```python
df
```

```python
df2 = pd.DataFrame(columns=['Long', 'Lat'])
```

```python
for i in df.index:
    loc = df.Location[i].split(", ")
    Long = loc[0].replace("(","")
    Lat = loc[1].replace(")","")
    
    tempdf = pd.DataFrame({"Long":[Long],
                           "Lat":[Lat]})
    df2 = df2.append(tempdf)
```

```python
df2.head(5)
```

```python
df2.Long = df2.Long.astype(float)
df2.Lat = df2.Lat.astype(float)
```

```python
def generateBaseMap(default_location=[df2.Long.values[0], df2.Lat.values[0]], default_zoom_start=11):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
base_map = generateBaseMap()
```

```python
df2.head(5)
```

```python
base_map
```

```python
HeatMap(data=df2[['Long', 'Lat']].groupby(['Long', 'Lat']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
```
