# This will create a string array of all of the possible combined names

```python
groupedNeighborhoods = []
for index, row in df_CD.iterrows():
    if "/" in row['CSA2010']:
        groupedNeighborhoods.append(df_CD['CSA2010'].loc[index])
```

# Seems like it worked.

# So we don't need to split them. Instead we need to change the Neighborhoods column in RPT

# My first thought was to split them in CD, duplicate the data.
# But then you would mess with the accuracy of the data
# So that won't work.

```python
print(groupedNeighborhoods)
```

```python
df_RPT.Neighborhood = df_RPT.Neighborhood.astype(str)
```

#  This algorithm will interate through the rows
# Then iterate through the array groupedNeighborhoods

# With these two indexes, we check if the Neighborhood lies in a specific group
# For example, Brooklyn is it's own thing in RPT. But in CD, it's listed as Brooklyn/Curtis Bay/Hawkins Point

# So, we're change the Neighborhood in RPT to the Group

# I had to specifically ignore "nan" because there was a group name that has "..nan.." in it, and it was getting picked up.

# I was already like 30mins into running it when I noticed. Such is life

```python
for index, row in df_RPT.iterrows():
    for i in groupedNeighborhoods:
        if (df_RPT.loc[index].Neighborhood in i) and (df_RPT.loc[index].Neighborhood != "nan"):
            #print(row['Neighborhood'])
            #print("Before: " + df_RPT.loc[index].Neighborhood)
            #df_RPT.replace(df_RPT.loc[index].Neighborhood, i
            df_RPT.loc[index, 'Neighborhood'] = i
            #print("After: " + df_RPT.loc[index].Neighborhood)
```
