```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = []
for index, row in df.iterrows():
    x.append([df.loc[index,'CityTax']])
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 1/3,
                                               random_state = 0)
```

```python
linearRegressor.fit(xTrain, yTrain)
```

```python
yPrediction = linearRegressor.predict(xTest)
```

# Noise

```python
plt.scatter(xTrain, yTrain, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.title('City Tax vs African American Population (Training)')
plt.xlabel('City Tax')
plt.ylabel('African American Population')
```

# You can see the neighhborhoods being visualized with those lines. Each one is a different neighborhood

```python
plt.scatter(xTest, yTest, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.title('City Tax vs African American Population (Training)')
plt.xlabel('City Tax')
plt.ylabel('African American Population')
```

```python
print('Coeifficients: ', linearRegressor.coef_)
print("Mean squared error: %.2f"
     % mean_squared_error(yTest, yPrediction))
print('Variance score: %.2f' 
     % r2_score(yTest, yPrediction))
```

# I need to trim this down. Maybe if I took the average or another estimator to give a CityTax average for each neighborhood

```python
df2 = pd.DataFrame(columns=['CityTaxAverage', 'paa10', 'pwhite10', 'pasi10',
                            'p2more10','ppac10', 'phisp10'])
```

```python
df2
```

```python
neighborhoods = df.Neighborhood.unique()
total = 0
count = 0
average = 0
arr = []
for i in neighborhoods:
    for index, row in df.iterrows():
        if i in row['Neighborhood']:
            total = row['CityTax'] + total
            count+=1
            aa = row['paa10']
            white = row['pwhite10']
            asi = row['pasi10']
            more = row['p2more10']
            pac = row['ppac10']
            hisp = row['phisp10']
            loc = row['Location']
            
    average = total/count
    #print(average)
    arr.append(average)
    
    tempdf = pd.DataFrame({"CityTaxAverage":[average],
                           "paa10":[aa],
                           "pwhite10":[white],
                           "pasi10":[asi],
                           "p2more10":[more],
                           "ppac10":[pac],
                           "phisp10":[hisp],
                           "Location":[loc]})
    df2 = df2.append(tempdf)
```
